```python
import pandas as pd
from sklearn.linear_model import LogisticRegression 
from sklearn.linear_model import SGDClassifier
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import Imputer
```

```python
COLUMNS = ['PassengerId', 'Survived', 'Pclass', 'Name', 'Sex', 'Age' ,'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']
```

```python
df_train = pd.read_csv('train.csv', skipinitialspace=True)
df_test = pd.read_csv('test.csv', skipinitialspace=True)
```

```python
df_train
```

```python
survived = df_train['Survived'].values
```

```python
age = df_train['Age'].values
```

```python
pclass = df_train['Pclass'].values
sex = df_train['Sex'].astype('category')
sex.
```

```python
training = np.column_stack((age, pclass))

imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
imp = imp.fit(training)

training = imp.transform(training)
training
```

## Define the Logistic Regression Model

```python
clf = SGDClassifier(loss="hinge", penalty="l2")
```

```python
clf.fit(training, survived)
```

## Prediction

```python
t_age = df_test['Age'].values
```

```python
t_pclass = df_test['Pclass'].values
```

```python
testing = np.column_stack((t_age, t_pclass))

imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
imp = imp.fit(testing)

testing = imp.transform(testing)
testing
```

```python
clf.predict(testing)
```
